```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "catalog_au")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "golden")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_automobile_data_transformed= spark.table(f"{catalogo}.{esquema_source}.automobile_data_transformed")
```

```python
df_transformed = df_automobile_data_transformed \
    .groupBy(col("body")) \
    .agg(
        max(col("wheel_base")).alias("max_wheel_base")
    ) \
    .orderBy(col("max_wheel_base").desc())
```

```python
df_transformed.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.automobile_data_partitioned")
```
